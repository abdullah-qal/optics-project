```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import j1  # Bessel function of first kind
```

```python
def airy_disk(intensity_size=500, aperture_radius=1.0, wavelength=550e-9, focal_length=1.0):
    x = np.linspace(-0.5e-6, 0.5e-6, intensity_size)
    y = np.linspace(-0.5e-6, 0.5e-6, intensity_size)
    X, Y = np.meshgrid(x, y)
    R = np.sqrt(X**2 + Y**2)

    k = 2 * np.pi / wavelength
    kr = k * aperture_radius * R / focal_length

    intensity = (2 * j1(kr) / kr)**2
    intensity[kr == 0] = 1.0  

    plt.figure(figsize=(6, 5))
    plt.imshow(intensity, cmap='inferno', extent=[x.min()*1e6, x.max()*1e6, y.min()*1e6, y.max()*1e6])
    plt.colorbar(label='Normalized Intensity')
    plt.title("Airy Disk Pattern (Wave Optics)")
    plt.xlabel("x (μm)")
    plt.ylabel("y (μm)")
    plt.show()
```

```python
airy_disk()
```
