```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def ray_optics_aberration():

    fig, ax = plt.subplots(figsize=(8, 4))

    lens_x = 0
    source_x = -4
    source_y = 1  

    for dy in np.linspace(-0.4, 0.4, 9):
        ray_y = source_y + dy
        lens_y = ray_y + 0.1 * ray_y**3  # simulate slight curvature effect
        image_x = 2 + 0.5 * lens_y**2    # rays miss perfect focus depending on height
        ax.plot([source_x, lens_x, image_x], [source_y, lens_y, 0], 'r')

    ax.axvline(x=lens_x, color='black', linestyle='--')
    ax.text(lens_x, 1.5, 'High-NA Lens', ha='center')

    ax.plot(source_x, source_y, 'ro')
    ax.text(source_x, source_y + 0.2, 'Off-Axis Point Source', ha='center')

    ax.set_title("Ray Optics: Aberrations in High-Aperture Lens")
    ax.set_xlim(-5, 5)
    ax.set_ylim(-2, 2)
    ax.set_aspect('equal')
    ax.set_xlabel('Optical Axis')
    ax.set_ylabel('Height')
    plt.grid(True)
    plt.show()

ray_optics_aberration()
```
