```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def ray_optics_perfect_focus():
    fig, ax = plt.subplots(figsize=(8, 4))
    lens_x = 0
    f = 2  
    source_x = -4
    source_y = 0

    for angle in np.linspace(-0.4, 0.4, 7):
        slope = np.tan(angle)
        lens_y = slope * (lens_x - source_x)
        image_x = lens_x + f
        image_y = 0
        ax.plot([source_x, lens_x, image_x], [source_y, lens_y, image_y], 'b')

    ax.axvline(x=lens_x, color='black', linestyle='--')
    ax.text(lens_x, 1.2, 'Lens', ha='center')

    ax.plot(source_x, source_y, 'ro')
    ax.text(source_x, -0.3, 'Point Source', ha='center')
    ax.plot(image_x, image_y, 'go')
    ax.text(image_x, -0.3, 'Image (Perfect Point)', ha='center')

    ax.set_title("Ray Optics: Perfect Image Formation (No Diffraction)")
    ax.set_xlim(-5, 5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.set_xlabel('Optical Axis')
    ax.set_ylabel('Height')
    plt.grid(True)
    plt.show()

ray_optics_perfect_focus()
```
